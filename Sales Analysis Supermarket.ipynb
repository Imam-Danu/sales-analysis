- <h4> Exploratory Data Analysis

```python
df['Order Date'].min()
```

```python
df['Order Date'].max()
```

```python
# Getting month year from dataset
df['month_year'] = df['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
# Grouping month year
df_trend = df.groupby('month_year').sum()['Sales'].reset_index()
```

```python
# Setting the figur size
plt.figure(figsize=(15,7))
plt.plot(df_trend['month_year'], df_trend['Sales'], color= 'red')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

- <h4> TOP 10 Product

```python
# Grouping product name column
prod_sales = pd.DataFrame(df.groupby('Product Name').sum() ['Sales'])
prod_sales
```

```python
top_prod_sales = prod_sales.sort_values('Sales',ascending=False)
#top 10 product by sales
top_prod_sales[:10]
```

- <h4> The Most Selling Products

```python
# Groupng product name
most_sell_prod = pd.DataFrame(df.groupby('Product Name').sum()['Quantity'])
```

```python
# Sorting most-sell_products
most_sell_prod=most_sell_prod.sort_values('Quantity', ascending=False)
# Top 10 selling products
most_sell_prod[:10]
```

- <h4> The Most Preferred Ship Mode

```python
# Setting Figure size
plt.figure(figsize = (10,8))

#plotting Ship Mode
sns.countplot(df['Ship Mode'])

plt.show()
```

- <h4> The Most Profitable Category and Sub Category

```python
# Grouping Category and Sub-Category
cat_subcat_profit = pd.DataFrame(df.groupby(['Category', 'Sub-Category']).sum()['Profit'])
```

```python
# Sorting the values
cat_subcat_profit.sort_values(['Category', 'Profit'], ascending = False)
```
